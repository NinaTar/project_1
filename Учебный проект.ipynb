## Введение

Нам представлен набор данных интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры. Данные собраны за период вплоть до конца 2016 года. В первую очередь загрузим данные и оценим их общее состояние: что они из себя представляют, имеются ли пропуски/дубликаты и как смотрятся визуально.

### Столбец year_of_release	

```python
data['year_of_release'].describe()
```

Хочу определить долю пропусков по столбцу в общем количестве значений.
В будущем мне это еще пригодится, поэтому решила написать функцию.

```python
def missed_values(column):
    print('Доля пропущенных значений составляет:', f"{(column.isna().sum()/len(column)):.2%}")
```

Но перед этим посмотрю что эти строки из себя представляют

```python
data['year_of_release'] = data['year_of_release'].astype(int)
```

###  Столбцы sales

Решила посмотреть столбцы sales вместе. Показатели финансовые - количество проданных копий, интересно визуально оценить их взаимосвязь.

```python
sales = data[['year_of_release','na_sales', 'eu_sales', 'jp_sales', 'other_sales']]
sales[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].describe()
```

```python
sales.query('year_of_release >= 2011')\
    .pivot_table(index='year_of_release', values=['na_sales', 'eu_sales', 'jp_sales',  'other_sales'], aggfunc='sum')\
    .plot(grid=True, style='o-')
plt.show()
```

Количество продаж падает из года в год. Сразу же посмотрим распределение долей по странам в продажах игр за последние 3 года (самый актуальный период из представленных данных).

```python
sales_pie = sales.query('year_of_release >= 2014')\
    .pivot_table(index='year_of_release', values=['na_sales', 'eu_sales', 'jp_sales', 'other_sales'], aggfunc='sum')

for ind in sales_pie.index:
    fig, ax = plt.subplots(1,1)
    sales_pie.loc[ind].plot(kind='pie', ax=ax, autopct='%1.1f%%', figsize=(4,4), \
                            title='Доли стран в общем количестве продаж по соответствующим годам')
```

```python
sales_pie
```


По продажам наблюдается интересная ситуация. Несмотря на то, что объём продаж игр стремительно падает, доли участия стран не сильно колеблются. Т.е.тенденции в компьютерной индустрии одинаково влияют на все рынки сбыта.
    


Также, рассматривая данный столбец, оценим имеющиеся данные в разрезе стран, в т.ч.распределение данных и наличие выбросов. Чтобы не дублировать код, создадим цикл по интересующим нас странам

```python
countries = ['na_sales','eu_sales', 'jp_sales']

for i in countries:
    sns.boxplot(x=sales.query('year_of_release >= 2014')[i])
    plt.xlim(0,1)
    plt.title(f'{i}, выбросы по продажам за период 2014-2016гг.')
    plt.show()
    
    print(sales.query('year_of_release >= 2014')[i].describe())
```


По всем рассмотренным странам мы наблюдаем одну и ту же картину - данные не сбалансированны, медиана сильно смещена к нулю, а стандартное отклонение выше средней в 2,5-4 раза в зависимости от страны. Всё говорит о наличие очень крупных выбросов, которые искажают общую статистику.

В столбце 'user_score' мы обнаружили необычную аббревиатуру - tbd. Предположительно она переводится как To Be Determined и подразумевает, что рейтинг по данной игре не был выставлен. ### Анализ платформ

Посмотрим объём продаж за 2016 год в разрезе платформ.

```python
plt.figure(figsize=(9,4))
ax = data.query('year_of_release==2016').groupby('platform')['total_sales'].sum()\
    .sort_values(ascending=False).plot(kind='bar', grid=True, ylim=(0,75), title='Объём продаж по платформам за 2016 год')
ax.bar_label(ax.containers[0], label_type='edge')
plt.xticks(rotation = 0)
ax.margins(y=0.5)
```

Построим ящики с усами для платформ по общим продажам также за последний год. Чтобы удобнее наблюдать все платформы, ограничим значения по оси y.

```python
plt.figure(figsize=(9,5))

ax = sns.boxplot(x=data.query('year_of_release>=2016')['platform'], y=data['total_sales'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.tight_layout()
plt.ylim(0,2)
plt.title('Общие продажи по платформам за 2016 год', size = 15)
plt.show()
```

Первые три позиции занимают платформы с достаточно неоднородными данными. Медианы их смещены вниз, т.к. в их данных присутствует много штучных крупных выбросов, которые влияют на общую картину. Любопытно отметить платформу WiiU, у которой максимально сбалансированные и равномерно распределенные данные. Визуально видно, что медиана находится почти посередине. 

Оценим устойчивость платформ за весь период и за последние пять лет, чтобы определить какие платформы имеют тенденцию к устойчивому спросу.

```python
plt.figure(figsize=(9,5))

sns.boxplot(x=data['year_of_release'], y=data['platform'].sort_values(ascending=True))
plt.tight_layout()
plt.grid(True)
plt.title('Жизненный цикл платформ за весь период')
plt.show()
```

```python
plt.figure(figsize=(8,4))
last_five_years_data = data.query('year_of_release>=2011')

sns.boxplot(x=last_five_years_data['year_of_release'], y=last_five_years_data['platform'].sort_values())
plt.tight_layout()
plt.title('Жизненный цикл платформ за последние 5 лет')
plt.grid(True)
```

### Анализ в разрезе оценок критиков и пользователей

Далее интересно посмотреть в разрезе самых популярных платформ на сколько оценки критиков и пользователей влияют на количество и суммы игровых покупок.

Создам две переменные critic_score_sales и user_score_sales. По ним соберу данные за последние три года по том-5 платформам. Посчитаю суммы покупок, их количество, медиану и выведу на график.

```python
critic_score_sales = data.query('year_of_release>=2014 and platform in ["PS4","PSV", "XOne", "PS3", "3DS"]')\
    .pivot_table(index='critic_score', values='total_sales', aggfunc=['sum','count', 'median'])

critic_score_sales.columns=['sum', 'count', 'median']

critic_score_sales.plot(grid=True, figsize=(11,4))
plt.title('Зависимость количества и суммы продаж от оценки критиков для топ-5 платформ за 2014-2016 гг.')
plt.show()

print(critic_score_sales.sort_values(by='sum', ascending=False).head(10))
```

```python
user_score_sales = data.query('year_of_release>=2014 and platform in ["PS4","PSV", "XOne", "PS3", "3DS"]')\
    .pivot_table(index='user_score', values='total_sales', aggfunc=['sum','count', 'median'])

user_score_sales.columns=['sum', 'count', 'median']

user_score_sales.plot(grid=True, figsize=(11,4))
plt.title('Зависимость количества и суммы продаж от оценки пользователей для топ-5 платформ за 2014-2016 гг.')
plt.show()

print(user_score_sales.sort_values(by='sum', ascending=False).head(10))
```


В целом, в обоих случаях по самым популярным платформам за период с 2014 по 2016 гг., мы можем определить пик по количеству и суммам покупок, который приходится на рейтинг 7-8.3 по оценкам пользователей, и 75-90 по оценкам критиков.
    
Стоит отметить, что высокий официальный рейтинг ведет за собой более высокую стоимость игры на платформе, в связи с чем спрос превалирует в среднем сегменте - комфортном соотношении цена/качество.
Если данные таблицы critic_score_sales отсортировать по медиане, мы увидем, что на рейтинг 90+ приходится самые "дорогие" игры, с малым числом продаж.
    
По пользовательским оценкам такой зависимости обнаружено не было.

## Портрет пользователя по регионам

### Платформы
Посмотрим предпочтения платформ по континентам, выведем по 5 самых популярных.

Чтобы не дублировать код для трех столбцов, создам функцию, которая соберет нужные данные и выведет график.

```python
def top_5 (data, index, column):
    ax = data.pivot_table(index=index , values=column, aggfunc='sum')\
    .sort_values(by=column, ascending=False).head()\
    .plot(kind='bar', grid=True, figsize=(6,3), alpha=0.7)
    
    ax.bar_label(ax.containers[0], label_type='edge')
    plt.xticks(rotation = 0)
    ax.margins(y=0.5)
```

```python
top_5(last_year_data, 'platform', 'na_sales')
plt.title('Топ-5 платформ по континентам за 2016 год:')
top_5(last_year_data, 'platform', 'eu_sales')
top_5(last_year_data, 'platform', 'jp_sales')
plt.show()
```

В Северной Америке и Европе все 5 мест заняли одинаковые платформы, разместившись на разных местах.

Хорошие продажы в суммарном исчислении можно отметить по платформам PS4, XOne и 3DS  во всех трех регионах.

### Жанры

Тоже сделаем и по жанрам

```python
top_5(last_year_data, 'genre', 'na_sales')
plt.title('Топ-5 жанров по континентам за 2016 год:')
top_5(last_year_data, 'genre', 'eu_sales')
top_5(last_year_data, 'genre', 'jp_sales')
plt.show()
```


По жанровым предпочтениям Япония также отличается от выбора Американских и Европейских пользователей. С крупным отрывом в Японии лидируют такие жанры как Action и Role-Playing. В других странах предпочтения между жанрами более сглаженные. Бесспорным лидером на рынке является жанр Shooter, за ним следуют Sports и Action.

### Рейтинг

Теперь посмотрим распределение среди рейтингов. На какую аудиторию рассчитаны самые популярные игры.

E - Рейтинг игр "для всех", 6+.

M -	рейтинг игр, рассчитанный на подростков.
